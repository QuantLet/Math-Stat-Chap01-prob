Illustrate the idea of probability

```python
import numpy as np

nSize = 20
# Generate 20 random numbers (0 or 1) with equal probability
random_numbers = np.random.choice([0, 1], size=nSize)

# Calculate cumulative sum to determine the number of 1s seen up to the current observation
cumulative_ones = np.cumsum(random_numbers)

# Calculate the probability of seeing a 1 up to the current observation
probabilities = cumulative_ones / np.arange(1, (nSize+1))

print(random_numbers)
print(probabilities)
```

```python
import matplotlib.pyplot as plt


# Plotting the probabilities of seeing a '1' up to the respective observations
plt.figure(figsize=(10, 10))
plt.plot(np.arange(1, (nSize+1)), probabilities, marker='o', linestyle='-')
plt.axhline(y=0.5, color='r', linestyle='--')  # Add a horizontal line at probability 0.5
plt.xlabel("Observation Number")
plt.ylabel("Probability of Seeing a '1'")
plt.title("Probability of Seeing a '1' Up to the Respective Observations")
plt.grid(True)
plt.xticks(np.arange(0, (nSize+1), 100))
plt.yticks(np.arange(0, 1.1, 0.1))
# plt.savefig('/content/drive/MyDrive/Colab Notebooks/output_Math_Stat/demoProb.png', transparent=True)
plt.show()
```
