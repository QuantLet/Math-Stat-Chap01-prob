Visualize Birthday problem

```python
def birthday_probability(n):
    """Compute the probability that at least 2 out of n people have the same birthday."""
    prob_no_shared_birthday = 1.0
    for i in range(n):
        prob_no_shared_birthday *= (365 - i) / 365

    return 1 - prob_no_shared_birthday


def main():
    num_people_list = list(range(1, 101))  # considering up to 100 people for demonstration
    probabilities = [birthday_probability(n) for n in num_people_list]

    # Store results in a list of tuples
    results = [(n, prob) for n, prob in zip(num_people_list, probabilities)]
    return results

# Execute the main function and store the results
birthday_results = main()
birthday_results[:10]  # Displaying the first 10 results for brevity
```

```python
import matplotlib.pyplot as plt

# Extracting data for plotting
num_people, probabilities = zip(*birthday_results)

# Plotting
plt.figure(figsize=(10, 10))
plt.plot(num_people, probabilities, marker='o', linestyle='-')
plt.xlabel("Number of People")
plt.ylabel("Probability")
plt.title("Probability of at least 2 people having a shared birthday vs Number of people")
plt.grid(True)
plt.axhline(y=0.5, color='r', linestyle='--')  # Add a horizontal line at probability 0.5
#plt.savefig('/content/drive/MyDrive/Colab Notebooks/output_Math_Stat/birthday.png', transparent=True)
plt.show()
```
